#  Variables are stored in both rows and columns

```python
weather = pd.read_csv("data/weather.csv")
weather
```

```python
weather_long =  weather.melt(id_vars=["id", "year", "month", "element"],
                            var_name="day",
                            value_name="temp")
```

```python
weather_long = weather_long.dropna()
```

```python
(weather_long
 .pivot_table(index=["id", "year", "month", "day"],
                        columns="element",
                        values="temp")
 .reset_index()
)
```
